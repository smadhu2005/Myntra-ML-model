```python
drop_list=['What is your favourite pattern to wear?','Which type of clothing do you buy most often?','What is your favorite clothing brand']
df=df.drop(columns=drop_list)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Data Cleaning
# Fill missing values
df = df.fillna('Unknown')

# Feature Extraction
# Label Encoding for Gender
label_encoder = LabelEncoder()
df['Please mention your gender'] = label_encoder.fit_transform(df['Please mention your gender'])
df['Age'] = label_encoder.fit_transform(df['Age'])

# One-Hot Encoding for categorical features
categorical_features = ['Do you follow current fashion trends?']

df = pd.get_dummies(df, columns=categorical_features)

# Display the processed data
print(df.head())
```

Predictions of Trending colors:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
# Define the target variable (favorite color to wear)
df['favorite_color'] = df['What is your favourite color to wear?(Any single color or color combinations)']

# Drop the original favorite color column
X = df.drop(columns=['What is your favourite color to wear?(Any single color or color combinations)', 'favorite_color'])
y = df['favorite_color']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
from sklearn.preprocessing import OneHotEncoder,StandardScaler
scaler = StandardScaler()
X1_train_stand = scaler.fit_transform(X_train)
X1_test_stand = scaler.transform(X_test)

rf_classifier = RandomForestClassifier(random_state=0)
rf_classifier.fit(X1_train_stand, y_train)
y_pred_rf = rf_classifier.predict(X1_test_stand)
```

```python
predictions = rf_classifier.predict(X1_test_stand)
```

```python
predictions
```

```python
predictions_df = pd.DataFrame(predictions, columns=['predicted_color'])

# Count the occurrences of each adaptability level
level_counts = predictions_df['predicted_color'].value_counts()

# Calculate percentage of each level
percentage_per_level = (level_counts / len(predictions_df)) * 100

print("Percentage of predicted color:")
print(percentage_per_level)
```
